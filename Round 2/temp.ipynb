```python
import pandas as pd
import numpy as np
import math 
import statistics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
prices = pd.read_excel('data/prices.xlsx')
prices['global_time'] = prices['day'] * 1000000 + prices['timestamp']
prices.sort_values(by='global_time', inplace=True)
prices
```

```python
unique_products = prices['product'].unique()

for product in unique_products:
    product_df = prices[prices['product'] == product]
    
    plt.figure(figsize=(10, 4))
    plt.plot(product_df['global_time'], product_df['mid_price'], marker='o')
    plt.title(f"Mid Price Time Series for Product: {product}")
    plt.xlabel("Global Time (Combined day & timestamp)")
    plt.ylabel("Mid Price")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
